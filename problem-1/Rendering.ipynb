```python
from overhang_detector import OverhangDetector, Mesh, mplot3d

from matplotlib import animation
```

```python
%matplotlib notebook
```

```python
zdetector = OverhangDetector()

zdetector.solveAndDisp(Mesh.from_file('tests/octahedron.stl').vectors, showFaces=False)[0].show()

fig, ax = zdetector.solveAndDisp(Mesh.from_file('tests/cube.stl').vectors)
fig.show()
```

```python
def init():
    #ax.scatter(xx, yy, zz, marker='o', s=20, c="goldenrod", alpha=0.6)
    return fig

def animate(i):
    ax.view_init(azim=i)
    return fig,

# # Animate
anim = animation.FuncAnimation(fig, animate, init_func=init,
                               frames=range(0,360,3), interval=100, blit=True)
# # Save
anim.save('cube.gif', fps=30)
```

```python
ydetector = OverhangDetector(down=(0, -1, 0))
fig2, ax2 = ydetector.solveAndDisp(Mesh.from_file('part.stl').vectors, showVertices=False)
fig2.show()
```

```python
def init():
    #ax.scatter(xx, yy, zz, marker='o', s=20, c="goldenrod", alpha=0.6)
    return fig2

def animate(i):
    ax2.view_init(azim=i)
    return fig2,

# # Animate
anim = animation.FuncAnimation(fig2, animate, init_func=init,
                               frames=range(0,360,2), interval=100, blit=True)
# # Save
anim.save('part.gif', fps=20)
```
